```python
#Ex 24: https://www.practicepython.org/exercise/2014/12/27/24-draw-a-game-board.html
#This exercise is Part 1 of 4 of the Tic Tac Toe exercise series. The other exercises are: Part 2, Part 3, and Part 4.

#Time for some fake graphics! Let’s say we want to draw game boards that look like this: 
'''

 --- --- --- 
|   |   |   | 
 --- --- ---  
|   |   |   | 
 --- --- ---  
|   |   |   | 
 --- --- --- 

This one is 3x3 (like in tic tac toe). Obviously, they come in many other sizes (8x8 for chess, 19x19 for Go, 
and many more).

Ask the user what size game board they want to draw, and draw it 
for them to the screen using Python’s print statement.

Remember that in Python 3, printing to the screen is accomplished by rint("Thing to show on screen")

Hint: this requires some use of functions, as were discussed previously on this blog and elsewhere on the Internet, 
like this TutorialsPoint link. http://www.tutorialspoint.com/python/python_functions.htm
'''
def board_output(game_type):
    
    #type(game_type)  
    if game_type == 'tic_tac_toe':
        print(" --- --- ---")
        print("|   |   |   |")
        print(" --- --- ---")
        print("|   |   |   |")
        print(" --- --- ---")
        print("|   |   |   |")
        print(" --- --- ---")
    
    elif game_type == 'chess':
        horizontal = (" ---")*8
        vertical = ("|   ")*9
        for element in range(8):
            print(horizontal)
            print(vertical)
        print(horizontal)
       
    elif game_type == 'go':
        horizontal = (" ---")*19
        vertical = ("|   ")*20
        for element in range(19):
            print(horizontal)
            print(vertical)
        print(horizontal)
        
    else: 
        print('No such board game exist')
   


def choose_boardtype():
    board_type = ('tic_tac_toe', 'chess', 'go')  #board_type[0] =TickTackToe, board_type[1]=Chess, board_type[2]=go
    
    try: 
        user_choice = int(input("Please choose 1 to play TickTacToe, choose 2 to play Chess, choose 3 to play go: "))
        if user_choice==1:
            board_output(board_type[0])
        elif user_choice==2: 
            board_output(board_type[1])
        elif user_choice==3:
            board_output(board_type[2])
        else:
            print("You did not choose between 1~3 for game type. Please play by the rule... Goodbye!")
    except Exception as e:
        print(e)
        print("You did not choose between 1~3 for game type. Please play by the rule... Goodbye!")
        
choose_boardtype()





```
